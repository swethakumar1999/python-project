# 3. Display the percentage distribution of the data in each year using the pie chart.


```python
df.groupby('year').count()[['state']].plot(y='state',kind='pie',figsize=(6,6),autopct='%1.1f%%',legend=False);
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.


```python
sns.displot(x=df['priceperlb'],kind='hist', bins = 20, kde = True)
plt.show()



#Using Histogram,Plot for 'priceperlb' and the distribution of the graph is Peak in somepoint.
#Where the 'priceperlb' value in between 1.3 to 1.5 and count value in between 50 to 60.
# So the graph is Positively Skewed.
```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
plt.scatter(x=df['numcol'],y=df['prodvalue'])
plt.title('numcol v/s prodvalue')
plt.show()


#The Relationship between 'numcol' vs 'prodvalue' is directly related with the representation of graph.
#The values of numcol and prodvalue is merged in somepoint.
#The graph values is also gradually increasing.
```

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.


```python
sns.catplot(data=df, x= 'year', y= 'prodvalue',kind='box',height=5,aspect=1.2)
plt.show()

#The Categorical variable 'year' and a numerical variable 'prodvalue' has a relationship and we find through boxplot,that it 
#has a outliers.
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
sns.pairplot(df[['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year']],corner = True);




#a)'numcol', 'totalprod' and 'prodvalue' are positively skewed.
#b)'yieldpercol' seems to be normally distributed.
#c)'totalprod' vs 'numcol' and 'totalprod' vs 'prodvalue' seems to be directly(positively) related to each other.
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python

sns.heatmap(data=df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr(),annot=True,cmap='YlGnBu');



#The correlation value between the columns of 'numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue'.
#Using heatmap the values have been predicted.
#Totalprod and numcol are Positively skewed.
#priceperlb and yieldpercol are Negatively skewed.
```
